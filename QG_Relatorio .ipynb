```python
import matplotlib.pyplot as plt
```

```python

# Suponha que 'qg' seja o seu DataFrame
qg_novaTabela = qg[['NMFILIAL', 'VALORVENDA']]

# Agrupe os dados por 'NMFILIAL' e calcule a soma dos 'VALORVENDA' para cada filial
agrupado = qg_novaTabela.groupby('NMFILIAL')['VALORVENDA'].sum()

# Crie o gráfico de barras
agrupado.plot(kind='bar')
plt.title('Total de Vendas por Filial')
plt.xlabel('Filial')
plt.ylabel('Total de Vendas')
plt.show()

#Salva o gráfico em um arquivo PNG (ou outro formato desejado)
plt.savefig('meu_grafico.png')
```
